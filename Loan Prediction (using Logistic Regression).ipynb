# Data Cleaning

```python
### Checking the missing values
```

```python
df.isnull().sum()
```

```python
# Filling in the missing values of 'Loan Amount' & 'Credit History' by the 'Mean' and 'Median' of the respective values
```

```python
df[df['LoanAmount'].isnull()]
```

```python
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].median())
```

```python
#### Dropping all missing values of other columns
```

```python
df.dropna(inplace = True)
```

```python
# checking if all the missing values have been dropped or not
df.isnull().sum()
```

```python
df.head(10)
```

```python
### Converting the categoical data to numerical data
```

```python
df['Loan_Status'] = df['Loan_Status'].map({'Y':1, 'N':0})
df['Loan_Status'].value_counts()
```

```python
df['Gender'] = df['Gender'].map({'Male':1,'Female':0})
df.Gender.value_counts()
```

```python
df['Married'] = df['Married'].map({'Yes':1,'No':0})
df.Married.value_counts()
```

```python
df['Dependents'] = df['Dependents'].map({'0':0,'1':1,'2':2,'3+':3})
df.Dependents.value_counts()
```

```python
df['Education'] = df['Education'].map({'Graduate':1,'Not Graduate':0})
df.Education.value_counts()
```

```python
df['Self_Employed'] = df['Self_Employed'].map({'Yes':1,'No':0})
df.Self_Employed.value_counts()
```

```python
df['Property_Area'] = df['Property_Area'].map({'Urban':2,'Semiurban':1, 'Rural':0})
df.Property_Area.value_counts()
```

```python
# Final Data frame
df.head(5)
```

```python
### Importing packages for classification algorithms
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
# splitting the data into train and test set
```

```python
X = df.iloc[1:504,1:12].values
y = df.iloc[1:504,12].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)
```

```python
# Logistic Reression
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)

lr_prediction = model.predict(X_test)
print('Logistic Regression Accuracy =', metrics.accuracy_score(lr_prediction, y_test) )
```
